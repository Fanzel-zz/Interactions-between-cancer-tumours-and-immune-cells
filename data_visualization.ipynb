```python
cell_counts = data['Tissue Category'].value_counts()
```

```python
plt.figure(figsize=(10, 6))
ax = cell_counts.plot(kind='bar', color='skyblue')
plt.title('Number of Cells by Tissue Category')
plt.xlabel('Tissue Category')
plt.ylabel('Number of Cells')
plt.xticks(rotation=45)

# Add quantity labels to each bar
for i in ax.patches:
    ax.text(i.get_x() + i.get_width()/2, i.get_height() + 0.3, str(int(i.get_height())), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
phenotype_counts = data['Phenotype'].value_counts()
clean_labels = phenotype_counts.index.str.replace(r'^(Triple|Double|Single)\_?(positive\_)?', '', regex=True)
phenotype_counts.index = clean_labels
phenotype_counts
```

```python
plt.figure(figsize=(10, 6))
ax = phenotype_counts.plot(kind='bar', color='skyblue')
plt.title('Cell Counts by Phenotype')
plt.xlabel('Phenotype')
plt.ylabel('Total Cells') 
plt.xticks(rotation=45)  #Rotate the x-axis labels 45 degrees
plt.grid(True, linestyle='--', alpha=0.6)  # add grid

#Add quantity labels to each bar
for i in ax.patches:
    ax.text(i.get_x() + i.get_width()/2, i.get_height() + 0.3, str(int(i.get_height())), ha='center', va='bottom')
plt.show()
```

```python
grouped_data = data.groupby(['Sample Name', 'Tissue Category']).size().reset_index(name='Cell Count')

# 如果想用堆叠条形图表示
# 转换数据以适用于堆叠条形图
pivot_data = grouped_data.pivot(index='Sample Name', columns='Tissue Category', values='Cell Count').fillna(0)
pivot_data.plot(kind='bar', stacked=True, figsize=(12, 8))
plt.title('Stacked Bar Chart of Cell Counts by Tissue Category')
plt.xlabel('Sample Name')
plt.ylabel('Cell Counts')
# plt.xticks(rotation=45)
plt.show()
```

```python
tissue_types = data['Tissue Category'].unique()
data['Phenotype'] = data['Phenotype'].astype(str)
data['Cell Type'] = data['Phenotype'].apply(lambda x: 'Tumor Cell' if 'CK+' in x else 'Immune Cell')
```

```python
sample_names = data['Sample Name'].unique()
num_samples = len(sample_names)
fig, axes = plt.subplots(nrows=num_samples, ncols=1, figsize=(8, 10*num_samples))


colors = {'Tumor Cell': 'red', 'Immune Cell': 'blue'}
markers = {'Immune Cell': '^','Tumor Cell': 'o'}

for ax, sample_name in zip(axes, sample_names):
    sample_data = data[data['Sample Name'] == sample_name]
    for cell_type in ('Tumor Cell', 'Immune Cell'):
        group = sample_data[sample_data['Cell Type'] == cell_type]
        ax.scatter(group['Cell X Position'], group['Cell Y Position'], 
                   c=colors[cell_type], marker=markers[cell_type], 
                   label=cell_type, alpha=0.1, edgecolors= None, s=20)
    ax.set_title(f'Spatial Distribution for {sample_name}')
    ax.set_xlabel('Cell X Position')
    ax.set_ylabel('Cell Y Position')
    ax.legend(title='Cell Type')
    ax.grid(True)

plt.tight_layout()
plt.show()
```

```python
phenotype_counts = data.groupby(['Tissue Category', 'Phenotype']).size().reset_index(name='Count')

# Convert counts to percentages within each tissue category
phenotype_counts['Percentage'] = phenotype_counts.groupby('Tissue Category')['Count'].transform(lambda x: 100 * x / x.sum())

# Plotting the results using a bar plot
plt.figure(figsize=(12, 8))
sns.barplot(data=phenotype_counts, x='Tissue Category', y='Percentage', hue='Phenotype')
plt.title('Percentage of Each Phenotype Across Tissue Categories')
plt.xlabel('Tissue Category')
plt.ylabel('Percentage')
plt.xticks(rotation=45)  # Rotate labels for better readability
plt.legend(title='Phenotype')
plt.show()
```

```python
count_data = data.groupby(['Sample Name', 'Phenotype']).size().unstack(fill_value=0)

# 绘制分组条形图
count_data.plot(kind='bar', figsize=(12, 8))

plt.title('Cell Count by Phenotype for Each Sample')
plt.xlabel('Sample Name')
plt.ylabel('Cell Count')
plt.legend(title='Phenotype')
plt.grid(True)
plt.show()
```

```python
count_data = data.groupby(['Annotation ID', 'Phenotype']).size().unstack(fill_value=0)

# 对细胞计数进行对数转换，使用log1p对0值安全
count_data_log = np.log1p(count_data)
count_data_log
```

```python
num_phenotypes = len(count_data_log.columns)

# 创建足够的子图
fig, axes = plt.subplots(num_phenotypes, 1, figsize=(12, 2 * num_phenotypes), sharex=True)

# 如果只有一个表型，确保axes是可迭代的
if num_phenotypes == 1:
    axes = [axes]

# 逐个绘制KDE图
for ax, phenotype in zip(axes, count_data_log.columns):
    sns.kdeplot(count_data_log[phenotype], fill=True, bw_adjust=0.5, ax=ax, label=f'{phenotype}')
    ax.set_title(f'Log-scaled KDE of {phenotype} Cell Counts')
    ax.set_ylabel('Density')
    ax.legend(loc='upper right')  # 将图例放在每个子图的右上角

# 只在最下面的子图上设置x轴标签
axes[-1].set_xlabel('Log-scaled Cell Count')

# 设置整体标题并调整布局
fig.suptitle('Log-scaled KDE of Cell Counts by Phenotype Across Samples', y=1.02)
plt.tight_layout()  # 紧凑布局
plt.show()
```

```python
p = sns.pairplot(count_data_log, diag_kind='kde')
p.map_lower(sns.kdeplot, levels=4, color=".2")
```

```python
clean_phenotype = count_data_log.columns.str.replace(r'^(Triple|Double|Single)\_?(positive\_)?', '', regex=True)
count_data_log.columns = clean_phenotype
```

```python
melted_data = count_data_log.reset_index().melt(id_vars=["Annotation ID"], var_name="Phenotype", value_name="Log-Scaled Cell Count")

plt.figure(figsize=(12, 8))
sns.violinplot(x='Phenotype', y='Log-Scaled Cell Count', data=melted_data, inner=None, scale="width", linewidth=0.5)
sns.boxplot(x='Phenotype', y='Log-Scaled Cell Count', data=melted_data, width=0.2, linewidth=2, fliersize=3, boxprops={'facecolor':'none', "zorder":10})
plt.xticks(rotation=45)
plt.title('Distribution of Log-Scaled Cell Counts by Phenotype')
plt.xlabel('Phenotype')
plt.ylabel('Log-Scaled Cell Count')
plt.show()
```

```python
total_cells_per_sample = data.groupby('Sample Name').size()

# 计算每个样本中每种表型的细胞数
phenotype_counts = data.groupby(['Sample Name', 'Phenotype']).size().unstack(fill_value=0)

# 计算表型占总细胞数的比例
phenotype_fractions = phenotype_counts.div(total_cells_per_sample, axis=0)

# 绘制每种表型的比例变化
phenotype_fractions.plot(kind='bar', figsize=(12, 8))
plt.title('Fraction of Cell Phenotypes per Sample')
plt.xlabel('Sample Name')
plt.ylabel('Fraction of Cells')
plt.legend(title='Phenotype')
plt.show()
```

The y-axis represents cell number/total cell number，so even if the total cell number is different, the relative distribution of phenotypes can be seen by comparing between samples.

### Distance from Tissue Category Edge (microns)

```python
plt.figure(figsize=(10, 6))

for phenotype in data['Phenotype'].unique():
    subset = data[data['Phenotype'] == phenotype]
    sns.kdeplot(data=subset, x='Distance from Tissue Category Edge (microns)',
                label=phenotype, fill=False)

plt.title('Contour Plot of Distance from Tissue Category Edge')
plt.xlabel('Distance from Tissue Category Edge (microns)')
plt.ylabel('Density')
plt.legend(title='Phenotype', title_fontsize='13', fontsize='11')
plt.grid(True)
```
